Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots. # Scatter plot
- used to undersatnd the relationship between two numerical entities

using iris dataset : all data set can be obtained from the keggel.

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

Scatter plot

```python
sns.scatterplot(x = 'petal_length', y = 'sepal_length', data = iris)
# sns.lable('Scatter plot')
plt.show()
```

Lets add the species

```python
sns.scatterplot(x = 'sepal_length',  y = 'petal_length', data = iris, hue = 'species')
plt.show()
```

```python
sns.scatterplot(x = 'sepal_length',  y = 'petal_length', data = iris, hue = 'species', style = 'species')
plt.show()
```

We can aslo use categorical column to determmine the color

```python
sns.scatterplot(x = 'sepal_length',  y = 'petal_length', data = iris, hue = 'petal_length', style = 'species')
plt.show()
```

# Distribution plot
- This is along with the frequency curve along with histogram

lets remove the histogram

```python
sns.distplot(diamond_df['price'], hist = False)
plt.show()
```

Let add color attribute to it.

```python
sns.distplot(diamond_df['price'], color = 'green')
plt.show()
```

# Note:

  1. Both frequencey curve and histogram

  3. only histogram

# Jin plot
 - Cobination of
   1. Scatter plot
   2. histogram

```python
# we are using iris data set
iris.head()
```

```python
#joint plot
sns.jointplot(x = 'sepal_length', y = 'petal_length', data = iris)
plt.show()
```

# Reading of joint plot
- The Above plot are combination of two sepeal lengths histogram on top and on right peral_length histogram
- and on middle we have scatter plot

```python
#joint plot
sns.jointplot(x = 'sepal_length', y = 'petal_length', data = iris, color = 'grey')
plt.show()
```

# Regration lines on hitogram and scatter plot
- use the kind = reg

```python
sns.jointplot(x = 'sepal_length', y = 'petal_length', data = iris,  color = 'olive', kind= 'reg')
plt.show()
```
